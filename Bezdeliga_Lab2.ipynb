#### Бинарный поиск

```python
def BinarySearch(l, tar):
    first = 0
    last = len(l)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if l[mid] == tar:
            index = mid
        else:
            if tar<l[mid]:
                last = mid -1
            else:
                first = mid +1
    return index
time_start = time.perf_counter()
f = BinarySearch(l, -64)
time_end = time.perf_counter()
print(l,"\n",f, f"\nБинарный поиск занял {time_end - time_start:0.10f} секунд")
```

```python
class node:
    # Инициализация узла. Значения левого ребенка, правого ребенка и родителя изначально заданы None, т.к. их может не быть вовсе.
    def __init__(self, value=None):
        self.value = value
        self.left_child = None
        self.right_child = None
        self.parent = None


class binary_search_tree:
    # Инициализация бинарного дерева. Корня изначально нет.
    def __init__(self):
        self.root = None

    def insert(self, value):
        # Когда корня нет - создать новый корень
        if self.root == None:
            self.root = node(value)
        else:
            self._insert(value, self.root)

    def _insert(self, value, cur_node):
        # cur node - текущий узел
        # Если значение меньше текущего узла
        if value < cur_node.value:
            # Если нет левого ребенка
            if cur_node.left_child == None:
                # Установка левого ребенка
                cur_node.left_child = node(value)
                # Установка родителя для левого ребенка
                cur_node.left_child.parent = cur_node
            else:
                # Рекурсивно вызываем эту функцию для левого ребенка
                self._insert(value, cur_node.left_child)
        # Если значение больше текущего узла
        elif value > cur_node.value:
            # Если нет правого ребенка
            if cur_node.right_child == None:
                # Установка правого ребенка
                cur_node.right_child = node(value)
                # Установка родителя для правого ребенка
                cur_node.right_child.parent = cur_node
            else:
                # Рекурсивно вызываем эту функцию для правого ребенка
                self._insert(value, cur_node.right_child)

    # Поиск элемента в дереве
    def search(self, value):
        # Если есть корень
        if self.root != None:
            return self._search(value, self.root)
        else:
             return False

    # Поиск элемента в дереве
    def _search(self, value, cur_node):
        # Если значение совпадает с текущим значением
        if value == cur_node.value:
            return True
        # Если значение меньше текущего значения и есть левый ребенок
        elif value < cur_node.value and cur_node.left_child != None:
            # Рекурсивно вызывает поиск для левого ребенка
            return self._search(value, cur_node.left_child)
        # Если значение больше текущего значения и есть правый ребенок
        elif value > cur_node.value and cur_node.right_child != None:
            # Рекурсивно вызывает поиск для правого ребенка
            return self._search(value, cur_node.right_child)
        return False

    def delete_value(self, value):
        global array
        global tree
        while value in array:
            array.remove(value)
        new_tree = binary_search_tree()
        for number in array:
            new_tree.insert(number)
        tree = new_tree

# Время старта
start = time.time()

tree = binary_search_tree()
for number in l:
    tree.insert(number)
# Время после построения дерева
tree_time = time.time() - start

tree.search(0)
# Время после поиска
first = time.time() - start

# Время на построение и поиск нулевого элемента
print("Время построения дерева и поиска нулевого элемента:", first)
```

#### Фибонначиев поиск

```python
def fibonacci_search(l, tar):
    size = len(l)
    start = -1
    f0 = 0
    f1 = 1
    f2 = 1
    while(f2 < size):
        f0 = f1
        f1 = f2
        f2 = f1 + f0
    while(f2 > 1):
        index = min(start + f0, size - 1)
        if l[index] < tar:
            f2 = f1
            f1 = f0
            f0 = f2 - f1
            start = index
        elif l[index] > tar:
            f2 = f0
            f1 = f1 - f0
            f0 = f2 - f1
        else:
            return index
    if (f1) and (l[size - 1] == tar):
        return size - 1
    return None
time_start = time.perf_counter()
f=fibonacci_search(l, -52)
time_end = time.perf_counter()
print(l,"\n",f, f"\nПоиск Фибоначчи занял {time_end - time_start:0.10f} секунд")
```

#### Интерполяционный поиск

```python
def Interpolation_search(l, tar):
    low = 0
    high = (len(l) - 1)
    while low <= high and tar >= l[low] and tar <= l[high]:
        index = low + int(((float(high - low) / ( l[high] - l[low])) * ( tar - l[low])))
        if l[index] == tar:
            return index
        if l[index] < tar:
            low = index + 1;
        else:
            high = index - 1;
    return -1
time_start = time.perf_counter()
f=Interpolation_search(l, -52)
time_end = time.perf_counter()
print(l,"\n",f, f"\nИнтерполяционный поиск занял {time_end - time_start:0.10f} секунд")
```

#### Простое рехэширование 

```python
class EasyRehash:
    def __init__(self, data):
        self.data = data.split(" ")
        self.array = [None] * len(self.data)
        self.size = len(self.data)

    def hash(self, array):
        s = 0
        for number in range(len(array)):
            s = s + ord(array[number])
        return s % self.size

    def _rehash(self, pos, i):
        return (pos + i) % self.size

    def rehash(self):
        for char_number in range(self.size):
            i = 0
            if self.array[self.hash(self.data[char_number])] == None:
                self.array[self.hash(self.data[char_number])] = self.data[char_number]
            else:
                while self.array[self._rehash(self.hash(self.data[char_number]), i)] != None:
                    i += 1
                self.array[self._rehash(self.hash(self.data[char_number]), i)] = self.data[char_number]

    def _search(self, element):
        return self.array.index(element)

    def search(self, element):
        if element in self.array:
            return self._search(element)
        else:
            print("Элемент {} не был хеширован".format(element))

    def _delete(self, number):
        self.size -= 1
        self.data.pop(number)
        self.array = [None] * self.size
        self.rehash()

    def delete(self, element):
        num = self.search(element)
        if type(num) == int:
            self._delete(num)

    def add(self, element):
        self.size += 1
        self.data.append(element)
        self.array = [None] * self.size
        self.rehash()
        
data = "a b c d e f g n r o b b"
print("Исходные данные:", data + "\n")
easy = EasyRehash(data)
easy.rehash()
print("Простое рехеширование:", easy.array)
easy.delete("b")
print("Удаление в простом рехэшировании:", easy.array)
easy.add("m")
print("Добавление в простом рехэшировании:", easy.array)
```

#### Рехэширование с помощью пвсевдо случайных чисел

```python
class RandRehash(EasyRehash):
    def rehash(self):
        for char_number in range(self.size):
            i = 0
            if self.array[self.hash(self.data[char_number])] == None:
                self.array[self.hash(self.data[char_number])] = self.data[char_number]
            else:
                while self.array[self._rehash(self.hash(self.data[char_number]), i)] != None:
                    i = random.randint(1, self.size)
                self.array[self._rehash(self.hash(self.data[char_number]), i)] = self.data[char_number]
                
data = "a b c d e f g n r o b b"
print("Исходные данные:", data + "\n")
rand = RandRehash(data)
rand.rehash()
print("Рандомное рехеширование:", rand.array)
rand.delete("b")
print("Удаление в рандомном рехэшировании:", rand.array)
rand.add("m")
print("Добавление в рандомном рехэшировании:", rand.array)
```

#### Метод цепочек

```python
class ChainRehash:
    def __init__(self, data):
        self.data = data.split(" ")
        self.array = [None] * len(self.data)
        self.size = len(self.data)

    def hash(self, array):
        sum = 0
        for number in range(len(array)):
            sum = sum + ord(array[number])
        return sum % self.size

    def rehash(self):
        for i in range(self.size):
            if self.array[self.hash(self.data[i])] == None:
                self.array[self.hash(self.data[i])] = self.data[i]
            elif self.array[self.hash(self.data[i])] != None:
                prevValue = self.array[self.hash(self.data[i])]
                if type(prevValue) != list:
                    self.array[self.hash(self.data[i])] = [prevValue]
                self.array[self.hash(self.data[i])].append(self.data[i])
        return self.array

    def find_value(self, key):
        for number, element in enumerate(self.array):
            if (type(element) is list):
                for number2, element2 in enumerate(self.array[number]):
                    if (element2 == key):
                        return [number, number2]
            else:
                if (element == key):
                    return [number]
        print("Элемент {} не был хеширован".format(key))

    def delete(self, element):
        lst = self.find_value(element)
        if type(lst) != None:
            self.data.remove(element)
            self.size -= 1
            self.array = [None] * self.size
            self.rehash()

    def add(self, element):
        self.data.append(element)
        self.size += 1
        self.array = [None] * self.size
        self.rehash()

data = "a b c d e f g n r o b b"
print("Исходные данные:", data + "\n")
chain = ChainRehash(data)
chain.rehash()
print("Рехеширование методом цепочек:", chain.array)
chain.delete("b")
print("Удаление в рехешировании методом цепочек:", chain.array)
chain.add("m")
print("Добавление в рехешировании методом цепочек:", chain.array)
```
