#### Бинарный поиск

```python
from random import randint
import time
l = sorted([randint(-100,100) for i in range (10)])
print (l)
```

```python
def BinarySearch(l, tar):
    first = 0
    last = len(l)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if l[mid] == tar:
            index = mid
        else:
            if tar<l[mid]:
                last = mid -1
            else:
                first = mid +1
    return index
time_start = time.perf_counter()
f = BinarySearch(l, -52)
time_end = time.perf_counter()
print(l,"\n",f, f"\nБинарный поиск занял {time_end - time_start:0.10f} секунд")
```

```python
def fibonacci_search(l, tar):
    size = len(l)
    start = -1
    f0 = 0
    f1 = 1
    f2 = 1
    while(f2 < size):
        f0 = f1
        f1 = f2
        f2 = f1 + f0
    while(f2 > 1):
        index = min(start + f0, size - 1)
        if l[index] < tar:
            f2 = f1
            f1 = f0
            f0 = f2 - f1
            start = index
        elif l[index] > tar:
            f2 = f0
            f1 = f1 - f0
            f0 = f2 - f1
        else:
            return index
    if (f1) and (l[size - 1] == tar):
        return size - 1
    return None
time_start = time.perf_counter()
f=fibonacci_search(l, -52)
time_end = time.perf_counter()
print(l,"\n",f, f"\nПоиск Фибоначчи занял {time_end - time_start:0.10f} секунд")
```

```python
def Interpolation_search(l, tar):
    low = 0
    high = (len(l) - 1)
    while low <= high and tar >= l[low] and tar <= l[high]:
        index = low + int(((float(high - low) / ( l[high] - l[low])) * ( tar - l[low])))
        if l[index] == tar:
            return index
        if l[index] < tar:
            low = index + 1;
        else:
            high = index - 1;
    return -1
time_start = time.perf_counter()
f=Interpolation_search(l, -52)
time_end = time.perf_counter()
print(l,"\n",f, f"\nИнтерполяционный поиск занял {time_end - time_start:0.10f} секунд")
```
